#### Streaming the file download without a progress bar

```python
import requests 

r = requests.get('https://cdn.pixabay.com/download/audio/2023/04/06/audio_6d1202b050.mp3?filename=hars-up-rise-145493.mp3',stream = True) 

total_expected_bytes = int(r.headers["Content-Length"]) 
```

```python
print(total_expected_bytes) 
```

```python
bytes_received = 0 

with open('sound.mp3','wb') as f:
    for chunk in r.iter_content(chunk_size = 128):
        f.write(chunk)
        print(f'{bytes_received} out of {total_expected_bytes}')
        bytes_received += 128 
```

```python
import requests 
from tqdm import tqdm

r = requests.get('https://download-cdn.jetbrains.com/python/pycharm-edu-2022.2.2.exe?_ga=2.242126208.320615278.1682231716-1137811871.1682231716&_gl=1*13uw8qh*_ga*MTEzNzgxMTg3MS4xNjgyMjMxNzE2*_ga_9J976DJZ68*MTY4MjIzMTcxNS4xLjEuMTY4MjIzMTc3OC4wLjAuMA..',stream = True) 

total_expected_bytes = int(r.headers["Content-Length"]) 

bytes_received = 0 

# create a progress bar 
progress_bar = tqdm(total = total_expected_bytes, unit='iB', unit_scale = True) 

with open('jetbrains-pycharm.exe','wb') as f:
    for chunk in r.iter_content(chunk_size = 128):
        progress_bar.update(len(chunk))
        f.write(chunk)
        bytes_received += len(chunk)

progress_bar.close()
```

```python
import requests 
from tqdm import tqdm

# Make a GET request to the download URL and stream the response data
r = requests.get('https://download-cdn.jetbrains.com/python/pycharm-edu-2022.2.2.exe?_ga=2.242126208.320615278.1682231716-1137811871.1682231716&_gl=1*13uw8qh*_ga*MTEzNzgxMTg3MS4xNjgyMjMxNzE2*_ga_9J976DJZ68*MTY4MjIzMTcxNS4xLjEuMTY4MjIzMTc3OC4wLjAuMA..', stream=True) 

# Get the total expected bytes from the response headers
total_expected_bytes = int(r.headers["Content-Length"]) 

bytes_received = 0 

# Create a progress bar with the total expected bytes and the unit of measurement
# Here, we're using 'iB' for "International System of Units" bytes
progress_bar = tqdm(total=total_expected_bytes, unit='iB', unit_scale=True) 

# Open a file for writing the downloaded content
with open('jetbrains-pycharm.exe', 'wb') as f:
    # Iterate over the response content in chunks and write each chunk to the file
    for chunk in r.iter_content(chunk_size=128):
        # Update the progress bar with the length of the current chunk
        progress_bar.update(len(chunk))
        f.write(chunk)
        bytes_received += len(chunk)

# Close the progress bar
progress_bar.close()
```
